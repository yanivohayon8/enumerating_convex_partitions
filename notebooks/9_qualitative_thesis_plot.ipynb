```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
means = np.array([3.000, 17.041, 92.52, 531.880, 3417.880, 19771.560])
std = np.array([0.0, 4.598, 17.808, 135.678, 825.990, 4826.123])

num_of_interior = np.arange(len(means)) + 3

# Set figure size (width, height in inches)
plt.figure(figsize=(10, 6))

# Plot the original data with error bars
ax = plt.subplot()
ax.errorbar(num_of_interior, means, yerr=std, marker="x", linestyle="", markeredgecolor="black", label="Observed Data")

# Exponential fit function
a = 0.0151
b = 1.7608

# Evaluate the fitted exponential function at the x values
start_sampling = 0
end_sampling = 8 
sampling_points = np.linspace(start_sampling,end_sampling)

fitted_values = a * np.exp(b * sampling_points)

# Plot the exponential fit
ax.plot(sampling_points, fitted_values, 'r-', label=f'Exponential Fit\ny = {a:.4f} * exp({b:.4f} * x)', linewidth=2)

# Set labels with bold font
ax.set_xlabel("Number of seed hull points")#, fontweight='bold')
ax.set_ylabel("Number of computed partitions")#, fontweight='bold')
ax.set_xticks(np.arange(9))
ax.set_xlim((0,9))
ax.set_ylim((0,27000))


# Add a legend
ax.legend(loc="upper left")

# Show the plot
plt.show()
```
